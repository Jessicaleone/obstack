```python
import pandas as pd
from elasticsearch import Elasticsearch
import json
from datetime import datetime
```

```python
x = [1,2,3,4,5]
y = [11,12,13,14,15]
l = [{'key': x[i], 'value': y[i]} for i in range(len(x))]
l
```

```python
es = Elasticsearch(timeout=5000)

page = es.search(index='jaeger-span-2022-05-19', scroll='10m', size=10000, body={
  "query":{
    "bool": {
      "must": [
          {"match": {"kind":"SERVER"}},
            {"match": {"name.keyword":"ts-auth-service_getToken"}},
          {
        "script": {
          "script": {
            "inline": "doc['traceID.keyword'].value == doc['spanID.keyword'].value"

                  }
                } 
              }]
            }
        }})

query_request = []

sid = page['_scroll_id']
print(page['hits']['total']['value'])

while(int(page['hits']['total']['value']) > 0):
    print("Scrolling...", str(int(page['hits']['total']['value'])))
    query_request.append(page['hits']['hits'])
    page = es.scroll(scroll_id=sid, scroll='10m')
    page['hits']['total']['value'] = len(page['hits']['hits'])


print(len(query_request))

result = []

for i in query_request:
    for j in i:
        result.append(j['_source'])

print(len(result))
```

```python
df= pd.DataFrame(result)
df
```

```python
for i in df['startTimeMillis'].index:
    df.loc[i,'startTimeMillis'] = datetime.fromtimestamp(df.loc[i,'startTimeMillis']/1000)
df    
```

```python
timestamp1 = 1652979531
timestamp2 = 1652983145
start = datetime.fromtimestamp(timestamp1)
end = datetime.fromtimestamp(timestamp2)
print(start,end)
```

```python
mask = (df['startTimeMillis'] > start) & (df['startTimeMillis'] <= end)
mask
```

```python
filtered_df=df.loc[mask]
filtered_df
```

```python
import seaborn as sns
g = sns.FacetGrid(filtered_df, row='name',sharey=False, sharex=False)
g.map_dataframe(sns.histplot, 'duration',log_scale=True)
```

```python
import matplotlib.pyplot as plt
# set up figure & axes
fig, axes = plt.subplots(figsize=(20,20))

# drop sharex, sharey, layout & add ax=axes
axes = filtered_df.hist(column='duration',by='name', ax=axes, bins=10000)

for ax in axes.flatten():
    ax.set_xscale('log')

# set title and axis labels
fig.text(0.01, 0.01, '')
fig.text(0.01, 0.01, '', rotation='vertical')
```
